```python
import pandas as pd

# Read the text file into a pandas DataFrame with error handling
file_path = '2241.txt'

# Load the data and skip bad lines
data = pd.read_csv(file_path, sep=':', header=None, names=['email', 'password'], error_bad_lines=False)

# Display the loaded data
print("Loaded Data:")
print(data)
```

```python
# Clean the data (drop duplicates, handle missing values)
data.dropna(inplace=True)
data.drop_duplicates(inplace=True)

# Display the cleaned DataFrame
print("\nCleaned Data:")
print(data)
```

```python
from pymongo import MongoClient

# Function to store data in MongoDB
def store_in_mongodb(data):
    client = MongoClient('mongodb://localhost:27017/')
    db = client['leaked_data']
    collection = db['leaks']
    collection.insert_many(data.to_dict('records'))
    print("Data inserted into MongoDB")

# Store the cleaned data in MongoDB
store_in_mongodb(data)
```
